```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
years_mna = np.array([2017, 2018, 2019, 2020, 2021])
mna_deals = np.array([39, 38, 35, 23, 39])

# Creating the line chart
plt.figure(figsize=(10, 6))
plt.plot(years_mna, mna_deals, marker='o', linestyle='-', color='blue')
plt.title('M&A Trends in Medical Aesthetics (2017-2021)')
plt.xlabel('Year')
plt.ylabel('Number of M&A Deals')
plt.grid(True)
plt.show()
```

```python
# Data for Projected Market Growth and Untapped Market Potential
categories = ['Current Customers', 'Potential Customers']
sizes = [15, 80]  # Sizes in millions, representing the number of customers

# Normalize the sizes for better visibility in the chart
max_size = max(sizes)
scaled_sizes = [(s / max_size) * 1000 for s in sizes]

# Create the bubble chart
fig, ax = plt.subplots()
x_positions = range(len(categories))  # X positions for the scatter plot

# Scatter plot for creating a bubble chart
for x, size in zip(x_positions, scaled_sizes):
    ax.scatter(x, 0, s=size, label=f'{categories[x]}: {sizes[x]}M')  # Use scaled sizes

# Customize the plot
ax.set_title('Market Growth and Untapped Potential')
ax.legend(loc='upper right')
ax.set_xticks(x_positions)
ax.set_xticklabels(categories)
ax.set_yticks([])  # Hide y-axis
ax.set_ylim(-1, 1)  # Limit y-axis to maintain chart aesthetics
ax.set_xlim(-0.5, 1.5)  # Adjust the x-axis limits to center the bubbles

plt.show()



```

```python
# Data for the enhanced customer acquisition channels pie chart
labels = ['Search Engines', 'Direct Visits', 'Social Media', 'Other']
sizes = [50, 30, 10, 10]  # Assuming percentages based on typical traffic distributions
colors = ['#1E90FF', '#00BFFF', '#4682B4', '#87CEFA']  # More visually appealing shades of blue
explode = (0.1, 0, 0, 0)  # Only "explode" the first slice (Search Engines)

fig, ax = plt.subplots(figsize=(8, 6))
wedges, texts, autotexts = ax.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
                                  startangle=140, shadow=True, pctdistance=0.85)

# Draw a circle at the center of pie to make it look like a donut
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)

ax.set_title('Customer Acquisition Channels', fontsize=16)
plt.setp(autotexts, size=12, weight="bold", color="white")  # Set properties for autotexts
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the chart
plt.show()
```

```python
# Re-entering data after reset
years = np.array([2018, 2019, 2020, 2021, 2022, 2023])
ideal_image_market_share = np.array([4.5, 4.7, 4.9, 5.0, 5.1, 5.1])  # Hypothetical historical market share in %
industry_growth = np.array([8, 8.5, 9, 9.5, 10, 10.8])  # Industry growth projections
competitors_market_share = np.array([6, 5.8, 5.7, 5.5, 5.3, 5.1])  # Example competitor data

fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar chart for Ideal Image's market share
ax1.bar(years, ideal_image_market_share, color='#0057B8', label='Ideal Image Market Share', width=0.4)
ax1.set_xlabel('Year')
ax1.set_ylabel('Market Share (%)', color='#0057B8')
ax1.tick_params(axis='y', labelcolor='#0057B8')
ax1.set_ylim(0, 15)  # Set the limit for the y-axis to make the chart easier to read

# Line chart for industry growth with emphasis and annotation
ax2 = ax1.twinx()
line, = ax2.plot(years, industry_growth, color='#FF6347', marker='o', linestyle='-', linewidth=2, label='Industry Growth Projection')
ax2.set_ylabel('Industry Growth Rate (%)', color='#FF6347')
ax2.tick_params(axis='y', labelcolor='#FF6347')
ax2.set_ylim(0, 12)  # Adjust the secondary axis to properly show the growth rate

# Adding annotations to emphasize growth
for i, txt in enumerate(industry_growth):
    ax2.annotate(f'{txt}%', (years[i], industry_growth[i] + 0.5), textcoords="offset points", xytext=(0,10), ha='center')

# Adding competitor data with a slight offset
ax1.bar(years + 0.4, competitors_market_share, color='#009BDF', label='Competitor Market Share', width=0.4)

# Title and legend
fig.suptitle('Ideal Image Market Share vs. Industry Growth', fontsize=16)
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))

# Show the chart
plt.show()
```
