```python
# Step 1: Upload the CSV File
from google.colab import files
uploaded = files.upload()

# Step 2: Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Step 3: Load and verify the data
df = pd.read_csv('nfc_south_efficiency_2021_2024.csv')

# Check the column names
print("Column names:", df.columns)

# Ensure 'Year' column is correctly formatted
df['Year'] = df['Year'].apply(lambda x: int(x.split('-')[0]))  # Convert year to integer

# Verify the data
print(df.head())

# Select features for the model (considering all relevant features)
features = ['Year', 'Yards per Play', 'Yards Allowed per Play', 'QB Rating', 'WR Efficiency', 'RB Efficiency',
            'Coaching Impact', 'Schedule Strength', 'Roster Rating', 'Draft Pick Impact', 'Free Agency Gain/Loss',
            'Betting Lines and Odds', 'Player Injury History', 'Scheduling Trends']

X = df[features]
y_offense = df['Offensive Points per Game']
y_defense = df['Points Allowed per Game']

# Train-test split
X_train, X_test, y_train_offense, y_test_offense = train_test_split(X, y_offense, test_size=0.2, random_state=42)
X_train, X_test, y_train_defense, y_test_defense = train_test_split(X, y_defense, test_size=0.2, random_state=42)

# Train Random Forest Regressor for offense
rf_offense = RandomForestRegressor(n_estimators=100, random_state=42)
rf_offense.fit(X_train, y_train_offense)

# Train Random Forest Regressor for defense
rf_defense = RandomForestRegressor(n_estimators=100, random_state=42)
rf_defense.fit(X_train, y_train_defense)

# Predict for 2024-2025 (Year = 2024)
pred_year = np.array([[2024, 5.5, 5.5, 90, 90, 90, 90, 88, 87, 86, 85, 88, 86, 85]])  # Example values for features in 2024
pred_offense = rf_offense.predict(pred_year)
pred_defense = rf_defense.predict(pred_year)

print(f"Predicted Offensive Points per Game for 2024-2025: {pred_offense[0]}")
print(f"Predicted Points Allowed per Game for 2024-2025: {pred_defense[0]}")

# Adding predictions to the DataFrame for visualization
df_pred = pd.DataFrame({
    'Year': [2024],
    'Offensive Points per Game': [pred_offense[0]],
    'Points Allowed per Game': [pred_defense[0]]
})

df_combined = pd.concat([df, df_pred])

# Plotting the Offensive Points per Game
plt.figure(figsize=(10, 5))
plt.plot(df_combined['Year'], df_combined['Offensive Points per Game'], marker='o', label='Offensive Points per Game')
plt.xlabel('Year')
plt.ylabel('Points per Game')
plt.title('Offensive Points per Game (2021-2025)')
plt.legend()
plt.grid(True)
plt.show()

# Plotting the Points Allowed per Game
plt.figure(figsize=(10, 5))
plt.plot(df_combined['Year'], df_combined['Points Allowed per Game'], marker='o', label='Points Allowed per Game')
plt.xlabel('Year')
plt.ylabel('Points per Game')
plt.title('Points Allowed per Game (2021-2025)')
plt.legend()
plt.grid(True)
plt.show()
```
